# Task 4: Most Preferred Investment Avenue

Objective: Identify the most preferred investment avenue

```python
data.columns
```

### Analyze Investment Avenues

```python
data["Investment_Avenues"].unique()
```

```python
data["Investment_Avenues"].value_counts()
```

```python
gender_investment_counts = data.groupby("gender")["Investment_Avenues"].value_counts()

gender_investment_counts
```

```python
investment_avenues = data['Investment_Avenues'].value_counts()

# Extracting labels and sizes
labels = investment_avenues.index
sizes = investment_avenues.values
colors = ['lightcoral' , 'skyblue']

# Plotting the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
grouped_data = data.groupby('gender')['Investment_Avenues'].value_counts().unstack()

# Plotting the grouped bar chart
sns.set(style='whitegrid')
grouped_data.plot(kind='bar', stacked=True)

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Investment Avenues by Gender')
plt.legend(title='Investment Avenue', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0, ha='right')
plt.show()
```

```python
data.groupby('Avenue')['Investment_Avenues'].value_counts()
```

```python
grouped_data = data.groupby('Avenue')['Investment_Avenues'].value_counts().unstack()

# Plotting the grouped bar chart
sns.set(style='whitegrid')
grouped_data.plot(kind='bar', stacked=True)

plt.xlabel('Avenue')
plt.ylabel('Count')
plt.title('Investment Avenues by Avenue')
plt.legend(title='Investment Avenues', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90, ha='right')
plt.show()
```

### Frequency Analysis

```python
investment_counts = data['Avenue'].value_counts()

# Determine the investment avenue with the highest frequency
max_investment_avenue = investment_counts.idxmax()

print("Investment avenue with the highest frequency:", max_investment_avenue)
```

# Task - 5: Reasons for Investment

Objective: Analyze and summarize reasons for investment choices


```python
data.head(10)
```

```python
print("Reasons for Equity:", data['Reason_Equity'].unique())
print("Reasons for Mutual:", data['Reason_Mutual'].unique())
print("Reasons for Bonds:", data['Reason_Bonds'].unique())
print("Reasons for FD:", data['Reason_FD'].unique())
```

```python
reason_equity = data['Reason_Equity'].value_counts()
reason_mutual = data['Reason_Mutual'].value_counts()
reason_bonds = data['Reason_Bonds'].value_counts()
reason_fd = data['Reason_FD'].value_counts()

# Set a color palette
sns.set_palette("pastel")

# Plotting pie charts
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # Creating a 2x2 grid of subplots

# Plot pie chart for Reason for Equity
axs[0, 0].pie(reason_equity, labels=reason_equity.index, autopct='%1.1f%%', startangle=140)
axs[0, 0].set_title('Reasons for Equity' , fontweight='bold')

# Plot pie chart for Reason for Mutual
axs[0, 1].pie(reason_mutual, labels=reason_mutual.index, autopct='%1.1f%%', startangle=140)
axs[0, 1].set_title('Reasons for Mutual' , fontweight='bold')

# Plot pie chart for Reason for Bonds
axs[1, 0].pie(reason_bonds, labels=reason_bonds.index, autopct='%1.1f%%', startangle=140)
axs[1, 0].set_title('Reasons for Bonds' , fontweight='bold')

# Plot pie chart for Reason for FD
axs[1, 1].pie(reason_fd, labels=reason_fd.index, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Reasons for FD' , fontweight='bold')


plt.tight_layout()
plt.show()
```
